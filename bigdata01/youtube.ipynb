```python
def get_information_from_response(response) :
    item_len = len(response.get('items'))
    _comments = {
        'published_date' : [response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('publishedAt') for i in range(item_len)],
        'updated_date' : [response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('updatedAt') for i in range(item_len)],
        'commentor' : [response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('authorDisplayName') for i in range(item_len)],
        'comment' : [response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('textOriginal') for i in range(item_len)],
        'like' : [response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('likeCount') for i in range(item_len)]
    }
    
    return pd.DataFrame(_comments)
```

```python
response = (
    youtube
    .commentThreads()
    .list(
      part = 'snippet,replies', 
      videoId = 'riMHgOa8yf4', 
      maxResults = 100
      )
    .execute()
    )

comments = get_information_from_response(response)

while 'nextPageToken' in response:
  response = (
    youtube
    .commentThreads()
    .list(
      part = 'snippet,replies',
      videoId = 'riMHgOa8yf4',
      pageToken=response.get('nextPageToken'),
      maxResults=100
      )
  .execute()
  )

  comments = comments.append(get_information_from_response(response))
```

```python

def get_replies_information_from_response(response) :
    comments = pd.DataFrame()

    for items in response.get('items') :
        if items.get('replies') :
            item_len = len(items.get('replies').get('comments'))
            
            _comments = {
                'published_date' : [items.get('replies').get('comments')[i].get('snippet').get('publishedAt') for i in range(item_len)],
                'updated_date' :[items.get('replies').get('comments')[i].get('snippet').get('updatedAt') for i in range(item_len)],
                'comment' :[items.get('replies').get('comments')[i].get('snippet').get('textOriginal') for i in range(item_len)],
                'commentor' :[items.get('replies').get('comments')[i].get('snippet').get('authorDisplayName') for i in range(item_len)],
                'like' :[items.get('replies').get('comments')[i].get('snippet').get('likeCount') for i in range(item_len)]
                }

            comments = comments.append(pd.DataFrame(_comments))

    return comments

```

```python
response = (
    youtube
    .commentThreads()
    .list(
      part = 'snippet,replies', 
      videoId = 'riMHgOa8yf4', 
      maxResults = 100
      )
    .execute()
    )

comments = get_information_from_response(response)
comments = comments.append(get_information_from_response(response))


while 'nextPageToken' in response:
  response = (
    youtube
    .commentThreads()
    .list(
      part = 'snippet,replies',
      videoId = 'riMHgOa8yf4',
      pageToken=response.get('nextPageToken'),
      maxResults=100
      )
  .execute()
  )

  comments = comments.append(get_information_from_response(response))
  comments = comments.append(get_replies_information_from_response(response))
```
