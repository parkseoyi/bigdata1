1~100 등수 순서 무작위로 나열 <br>
1~20등 => 1등급 <br>
21~40등 => 2등급 <br>
... <br>
80등~100등 => 5등급 <br>

````python
ranks = np.random.choice(np.arange(1,101), size=100)
````

````python
grades = []
for rank in ranks :
    if rank > 80 :
        grades.append(5)
    elif rank > 60 :
        grades.append(4)
    elif rank > 40 :
        grades.append(3)
    elif rank > 20 :
        grades.append(2)
    else :
        grades.append(1)
````

````python
ranks
````

````python
grades
````

````python
ranks.sort()
````

````python
ranks[:20] = 1
ranks[20:40] = 2
ranks[40:60] = 3
ranks[60:80] = 4
ranks[80:] = 5
````

````python
# 함수만들기
# input 돈
# output 몇장 당첨

from tqdm import tqdm

def lotto(money, win_numbers):
    numbers = np.random.randint(0,46,size=(money//1000, 6))
    win_count = 0

    for number in tqdm(numbers) :
        number.sort()

        if list(number) == win_numbers :
            win_count += 1
    
    return win_count
````
